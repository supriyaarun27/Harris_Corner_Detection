```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2 
image = cv2.imread('Desktop/Udacity/harris.jpg')
plt.imshow(image)
%matplotlib inline 
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
print(image_copy.shape)
```

```python
#convert image to grey scale 

grey= cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

#convert to float type 
grey = np.float32(grey)

#detect corners
dst = cv2.cornerHarris(grey, 2, 3, 0.04)
dst=cv2.dilate(dst, None) #to brighten the region 
plt.imshow(dst, cmap='gray')
```

```python
#disaply only stron corners 
thresh = 0.01*dst.max()

#create copy of the image to draw the corners on 
corner_image= np.copy(image_copy)

#iterate through the corners and draw it on the corner_image
for j in range(0, dst.shape[0]): 
    for i in range(0, dst.shape[1]):
        if(dst[j,i]>thresh): 
            cv2.circle(corner_image, (i,j), 2, (0,255,0), 1)

plt.imshow(corner_image)
                          
        
```
